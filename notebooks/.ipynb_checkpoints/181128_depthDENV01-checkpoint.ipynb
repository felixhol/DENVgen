## Read depth of single cell data

Get read depth of single cell data and store per cell in dataframes grouped by MOI/timepoint

```python
'''
author:     Felix Hol
date:       28 Nov 2018
content:    Synonymous vs. non-synonymous SNV analysis of DENV single cell data
'''

from pysam import VariantFile
import pandas as pd
import matplotlib.pyplot as plt
import os
import glob
import pysam
import numpy as np
import math
```

```python
baseDir = "/Users/felix/Documents/arbovirusGenetics/DENV_ZIKV_singelCell/pySamProcessing/depthfilesDENV01/"
os.chdir(baseDir)
#files = glob.glob('*.txt')
```

```python
depthThreshold = 9

files4 = glob.glob('*602*.txt')
files4_2 = glob.glob('*603*.txt')
files12 = glob.glob('*605*.txt')
files12_2 = glob.glob('*606*.txt')
files24 = glob.glob('*608*.txt')
files24_2 = glob.glob('*609*.txt')
files48 = glob.glob('*611*.txt')
files48_2 = glob.glob('*612*.txt')

depth_df4 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df12 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df24 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df48 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df4_2 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df12_2 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df24_2 = pd.DataFrame(columns = ['cellID', 'depth'])
depth_df48_2 = pd.DataFrame(columns = ['cellID', 'depth'])

allDepth4 = np.zeros(shape=(0,10723))
allDepth4_2 = np.zeros(shape=(0,10723))
allDepth12 = np.zeros(shape=(0,10723))
allDepth12_2 = np.zeros(shape=(0,10723))
allDepth24 = np.zeros(shape=(0,10723))
allDepth24_2 = np.zeros(shape=(0,10723))
allDepth48 = np.zeros(shape=(0,10723))
allDepth48_2 = np.zeros(shape=(0,10723))

_=fig=plt.figure(figsize=(12, 7), dpi=100)

numCells4 = 0
numCells12 = 0
numCells24 = 0
numCells48 = 0

numCells4_2 = 0
numCells12_2 = 0
numCells24_2 = 0
numCells48_2 = 0

for i in files4:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells4 = numCells4 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df4 = depth_df4.append(df2, ignore_index = True)
            plt.semilogy(f.depth,'g',alpha=0.1)
            allDepth4 = np.vstack([allDepth4, f.depth.values])

for i in files12:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells12 = numCells12 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df12 = depth_df12.append(df2, ignore_index = True)
            plt.semilogy(f.depth,'m',alpha=0.1)
            allDepth12 = np.vstack([allDepth12, f.depth.values])
            
for i in files24:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells24 = numCells24 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df24 = depth_df24.append(df2, ignore_index = True)
            plt.semilogy(f.depth,'r',alpha=0.1)
            allDepth24 = np.vstack([allDepth24, f.depth.values])

for i in files48:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells48 = numCells48 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df48 = depth_df48.append(df2, ignore_index = True)
#             plt.semilogy(f.depth,'b',alpha=0.1)
            allDepth48 = np.vstack([allDepth48, f.depth.values])

            
print('numCells4') 
print(numCells4)
print('numCells12') 
print(numCells12)
print('numCells24') 
print(numCells24)
print('numCells48') 
print(numCells48)
plt.semilogy(allDepth4.mean(axis=0),'k')
plt.semilogy(allDepth12.mean(axis=0),'k')
plt.semilogy(allDepth24.mean(axis=0),'k')
plt.semilogy(allDepth48.mean(axis=0),'k')
# plt.semilogy(allDepth48_2.mean(axis=0),'k')
plt.xlim([1,10750])

_=fig=plt.figure(figsize=(12, 7), dpi=100)
            
for i in files4_2:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells4_2 = numCells4_2 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df4_2 = depth_df4_2.append(df2, ignore_index = True)
            plt.semilogy(f.depth,'g',alpha=0.1)
            allDepth4_2 = np.vstack([allDepth4_2, f.depth.values])


for i in files12_2:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells12_2 = numCells12_2 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df12_2 = depth_df12_2.append(df2, ignore_index = True)
            plt.semilogy(f.depth,'m',alpha=0.1)
            allDepth12_2 = np.vstack([allDepth12_2, f.depth.values])
            
for i in files24_2:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells24_2 = numCells24_2 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df24_2 = depth_df24_2.append(df2, ignore_index = True)
#             plt.semilogy(f.depth,'r',alpha=0.1)
            allDepth24_2 = np.vstack([allDepth24_2, f.depth.values])            
            
            
for i in files48_2:
    if not os.stat(i).st_size == 0 :
        f = pd.read_csv(i, sep='\t', header=None)
        f.columns = ['genome', 'pos', 'depth']
        meanDepth = f["depth"].mean()
        cellID = str(i[5:len(i)-4])
        if meanDepth > depthThreshold:
            numCells48_2 = numCells48_2 +1
            df2 = pd.DataFrame([[cellID, meanDepth]], columns = ['cellID', 'depth'])
            depth_df48_2 = depth_df48_2.append(df2, ignore_index = True)
#             plt.semilogy(f.depth,'b',alpha=0.1)
            allDepth48_2 = np.vstack([allDepth48_2, f.depth.values])
            
print('numCells4_2') 
print(numCells4_2)
print('numCells12_2') 
print(numCells12_2)
print('numCells24_2') 
print(numCells24_2)
print('numCells48_2') 
print(numCells48_2)
plt.semilogy(allDepth4_2.mean(axis=0),'k')
plt.semilogy(allDepth12_2.mean(axis=0),'k')
plt.semilogy(allDepth24_2.mean(axis=0),'k')
plt.semilogy(allDepth48_2.mean(axis=0),'k')
# plt.semilogy(allDepth48_2.mean(axis=0),'k')
plt.xlim([1,10750])
```
